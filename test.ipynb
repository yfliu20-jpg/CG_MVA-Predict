```python
from seeq import spy
import pandas as pd
import numpy as np

# Schedule the current notebook to run every 5 minutes
spy.jobs.schedule('every 15 minutes')

# Get the current user's timezone from the session
user_tz = spy.utils.get_user_timezone(spy.session)

# Get the current timestamp
current_time = pd.Timestamp.now(tz=user_tz)

# Construct a DataFrame with just one data point (random data) indexed by the current time
data = pd.DataFrame({'random_data': [np.random.random()]}, index=[current_time])

# Print the DataFrame to check the data point
print("当前数据点：")
print(data)

# Push the data point to the specified Workbook and Worksheet in Seeq
push_results = spy.push(data=data, workbook='test', worksheet='test1')
print("推送结果：")
print(push_results)
```
